# REST
- REST == "REpresenational State Transfer"
- Resource-based
- Represenations
- Six Constraints:
- - Uniform interface
- - Stateless
- - Client-server
- - Cacheable
- - Layered System
- - Code on demand

## Resource Based
- Things vs. Actions
- Nouns(resource) vs. Verbs
- Versus SOAP-RPC
- Resources are Identified by URIs
- - Multiple URIs may refer to same resource
- Separate from their representations

## Representations
- How resources get manipulated
 Part of the resource state
 - - Transferred between client and server
 - Typically JSON or XML
 - Example:
 - - Resource: person (Todd)
 - - Service: contact information (GET)
 - - Representation:
 - - - name, address, phone number
 - - - JSON or XML format
 
 ## Uniform Interface
 - Defines the interface between client and server (It means we use the HTTP specs with URI's being our resource names and we use the HTTP verbs as the actions that we are going to take on those resources)
 - Simplifies and decouples the architecture
 - Fundamental to RESTful design
 - For us this means:
 - - HTTP vervs (GET, PUT, POST, DELETE)
 - - URIs (resource name)
 - - HTTP response (status, body)
 
 
 ## Stateless
 - Server contains no client state
 - Each request contains enough context to process the message
 - - Self-descriptive messages (Each message has enough information or context for the server to process that message)
 - Any session state is held on the client
 
There are occations where it is statefull, ex. Ouath, it keeps a state, but it is build in a RESTful manner and not technically stateful.

## Client-Server
- Assume a disconnected system
- Separation of concerns
- Uniform interface is the link between the two

## Cacheable
Server responses (representations) are cacheable
- Implicitly: a client caches the information
- Explicitly: the server caches the information
- Negotiated: client and server figures how long is the information cached

## Layered System
- Client can't assume direct connection to serve
- Software or hardware intermediaries between client and server
- Improves scalability

## Code On Demand
- Servers can temporarily extend client
- Transfer logic to client
- Client executes logic
- For example: 
- - Java applets
- - JavaScript
- The only optional constraint

## Summary
- Violating any constraint other than Code on Demand means service is not strictly RESTful
- - Example: Three-legged OAUTH2
- Compliance witht REST constraints allows:
- - Scalability
- - Simplicity
- Modifiability
- Visibility
- Portability
- Reliabiltiy
