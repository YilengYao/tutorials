### Foreign Key ONDELETE
When we delete a row on a parent node. there are 3 options:
- RESTRICT (NO ACTION)
- CASCADE
- SET NULL

#### RESTRICT
Default, prevent us to delete the parent, if you delete the parent row, you going to get an error

#### CASCADE
If you delete the parent row, the child row will be deleted as well.

#### SET NULL
If we delete the parent, the foreign key that reference the parent is set to null.

# Normalization
Normalization is a process of structuring a relational database to reduce data redundancy and improve data integrity.

Normalization has 3 main steps
- 1nf (first normal form)
- 2nf
- 3nf (Normalized)

A subsequent normal form need the previous normal form to be realized.


## 1NF
A relationship is in forst normal form if and only if the domain of wach attribute contains only atomic values, and the values of each attribute contains only a single value. form that domain.
- Atomic values(Singularity): each column only contains a single value, cannot contain multiple values.
- A column should contain values of the same type.
- each column should have unique name 

In 1NF, every column describes the key.

## 2NF
A relationship where 1NF is satisfied, and every non-key attributes is fully dependent on primary key. 2NF elimates partial dependencies on primary key.

Usually occurs in many to many relationships. 

If you have a table with composite primary keys, any column not belonging to the primary keys must be dependent on all the primary keys for uniqueness. If the column is dependent on a subset of the primary keys then it is partially dependent on the primay keys and is not 2NF.

## 3NF
3NF reduces the duplication of data, avoid data anomalies, ensure referential integrity, and simplify data management.

Conditions for 3NF:
- 2NF is statisfied
- Table only contains columns that are non-transitively dependent on the primary key.

Lets say that column A is the primary key. Column B is dependent on column A. Column C is dependent on column B. A <- B <- C. So C is transitively dependent on the primary key. Such a table is not 3NF.

To be 3NF column C must be directly depend on column A.

# Transaction

# ACID
A database transaction is an indivisible unit of work.

Either all steps of a transaction are successful or no steps happens at all.

So if any steps in a transaction is unsuccessful, the steps will be rollback and all partial transaction will be undone completely

ACID
- Atomic: All changes to the data must be performed successfully or not at all.
- Consistent: Data must be in a consisten state before and after the transaction.
- Isolated: No other process can change the data while the transaction is running
- Durable: The changes made by a transaction must persist.
